```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.aalto.fi/en/department-of-computer-science/faculty-0"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

FacultyData = []

members = soup.find_all('div', class_='aalto-profile-card')

for member in members:
    # Name
    name_tag = member.find('h3', class_='aalto-profile-card__name')
    name = name_tag.get_text(strip=True) if name_tag else "N/A"

    # Designation
    label_tag = member.find('div', class_='aalto-profile-card__label')
    designation = label_tag.get_text(strip=True) if label_tag else "N/A"

    # Email aur Phone
    email = "N/A"
    phone = "N/A"
    links = member.find_all('a', href=True)
    for link in links:
        if 'mailto:' in link['href']:
            email = link['href'].replace('mailto:', '')
        elif 'tel:' in link['href']:
            phone = link.get_text(strip=True)

    FacultyData.append({
        'Name': name,
        'Designation': designation,
        'Email': email,
        'Phone': phone
    })

df_faculty = pd.DataFrame(FacultyData)

print(f"Succesfully! Total {len(FacultyData)} members data has been fetched.")
display(df_faculty)
```
